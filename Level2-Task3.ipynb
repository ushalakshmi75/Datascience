# Feature Engineering

```python
import pandas as pd
#load the dataset
df=pd.read_csv('c:/Users/Usha/Downloads/dataset.csv.csv')
#display the first few rows to understand the dataset
print(df.head())
```

## Extract length of restaurant name and address

```python
if 'Restaurant Name' in df.columns:
    df['Restaurant_Name_Length']=df['Restaurant Name'].astype(str).apply(len)
if 'Address' in df.columns:
    df['Address_Length']=df['Address'].astype(str).apply(len)
```

## Encoding categorical variables

### Create new features like "Has Table Booking" or "Has Online Delivery"

```python
if 'Has Table booking' in df.columns:
    df['Has_Table_Booking']=df['Has Table booking'].map({'Yes':1,'No':0})
if 'Has Online delivery' in df.columns:
    df['Has_Online_Delivery']=df['Has Online delivery'].map({'Yes':1,'No':0})
```

## Display the transformed data

```python
print(df.head())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('c:/Users/Usha/Downloads/dataset.csv.csv')

# Display first few rows to understand the structure of the dataset
print("Dataset Preview:")
display(df.head())

# Extract additional features
df['Restaurant_Name_Length'] = df['Restaurant Name'].apply(lambda x: len(str(x)))
df['Address_Length'] = df['Address'].apply(lambda x: len(str(x)))

# Encoding categorical variables
# Assuming columns 'Has Table booking' and 'Has Online delivery' exist in the dataset
df['Has_Table_Booking'] = df['Has Table booking'].apply(lambda x: 1 if str(x).lower() in ['yes', 'true'] else 0)
df['Has_Online_Delivery'] = df['Has Online delivery'].apply(lambda x: 1 if str(x).lower() in ['yes', 'true'] else 0)

# Display processed dataset preview
display(df.head())

# Save the modified dataset
df.to_csv('c:/Users/Usha/Downloads/dataset.csv.csv', index=False)
print("Feature Engineering completed. Modified dataset saved as 'Modified_Dataset.csv'")
```
