# Descriptive Analysis

```python
#import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#load the dataset
df=pd.read_csv('https://drive.usercontent.google.com/download?id=1eS-q31uXdtdpSOg15g7SZ65rZjBM7gBO&export=download&authuser=0&confirm=t&uuid=9b80a293-3ecc-4384-9104-01f94bad3aed&at=AIrpjvPqQ2wonqOuzuiI__NUKLbD:1738586972677.csv')
```

### 1.Calculate basic statistical measures

```python
print("Basic Statistical Measures:")
print(df.describe())
```

### 2.Explore the distribution of Categorical variables

```python
#Country code distribution
plt.figure(figsize=(10,5))
sns.countplot(x=df['Country Code'])
plt.title("Distribution of Restaurants by Country Code")
plt.xticks(rotation=90)
plt.show()
```

```python
#Currency Distribution
plt.figure(figsize=(10,5))
sns.countplot(y=df['Currency'])
plt.title("Distribution of Restaurants by Currency")
plt.show()
```

### 3.Top 5 Cities with highest number of restaurants

```python
plt.figure(figsize=(10,5))
sns.countplot(x=df['City'],order=df['City'].value_counts().iloc[:5].index)
plt.title("Top 5 Cities with Highest Number Of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.show()
```

```python
print("Top 5 Cities With Highest Number of Restaurants:")
print(df['City'].value_counts().head(5))
```

### 4.Top 10 Cuisines with Highest Number of Restaurants

```python
plt.figure(figsize=(10,5))
sns.countplot(y=df['Cuisines'],order=df['Cuisines'].value_counts().iloc[:10].index)
plt.title("Top 10 cuisines with Highest Number of Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cuisines")
plt.show()
```

```python
print("Top 10 Cuisines with Highest Number of Restaurants:")
print(df['Cuisines'].value_counts().head(10))
```
