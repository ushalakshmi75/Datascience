### 1.percentage of restaurants offering table booking and online delivery

#### visualization for percentage of restaurants

```python
table_booking_pct=df["Has Table booking"].value_counts(normalize=True)*100
online_delivery_pct=df["Has Online delivery"].value_counts(normalize=True)*100
fig,axes=plt.subplots(1,2,figsize=(12,5))
axes[0].pie(table_booking_pct,labels=table_booking_pct.index,autopct='%1.1f%%',colors=['lightblue','orange'])
axes[0].set_title("Table Booking Availability")
axes[1].pie(online_delivery_pct,labels=online_delivery_pct.index,autopct='%1.1f%%',colors=['lightgreen','red'])
axes[1].set_title("Online Delivery Availability")
plt.show()
```

### 2. compare average ratings for restaurants with and without table booking

```python
avg_rating_with_table=df[df["Has Table booking"]=="yes"]["Aggregate rating"].mean()
avg_rating_without_table=df[df["Has Table booking"]=="no"]["Aggregate rating"].mean()
print(f"Average Rating (With Table Booking):{avg_rating_with_table:.2f}")
print(f"Average Rating (Without Table Booking):{avg_rating_without_table:.2f}")
#visualization
sns.barplot(x=["With Table Booking","Without Table Booking"],y=[avg_rating_with_table,avg_rating_without_table],palette=["red","yellow"])
plt.title("Average Rating Comparison")
plt.ylabel("Average Rating")
plt.show()
```

###visualization

```python
online_delivery_price=df.groupby("Price range")["Has Online delivery"].value_counts(normalize=True).unstack()*100
online_delivery_price.plot(kind="bar",stacked=True,figsize=(10,6),color=["pink","skyblue"])
plt.title("Online Delivery Availability by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Percentage(%)")
plt.legend(["No","Yes"])
plt.show()
```
